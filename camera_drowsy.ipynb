```python
import cv2
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import load_model
from playsound import playsound
import time

model = load_model('C:/Users/Mailizar Zein/Downloads/DrowsyData/drowsy_detection.h5')
alarm_sound_path = 'alarm path'

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    cap = cv2.VideoCapture(1)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

status = "focus"
percentage = 0
closed_eyes_duration = 0
closed_eyes_start_time = 0
alarm_duration_threshold = 3

while True:
    ret, frame = cap.read()

    resized_frame = cv2.resize(frame, (224, 224))
    resized_frame = np.expand_dims(resized_frame, axis=0)
    resized_frame = resized_frame / 255.0

    predictions = model.predict(resized_frame)

    if predictions > 0.6:
        status = "tidak focus"
        percentage = int(predictions * 100)

        if closed_eyes_start_time == 0:
            closed_eyes_start_time = time.time()
        else:
            closed_eyes_duration = time.time() - closed_eyes_start_time
            if closed_eyes_duration > alarm_duration_threshold:
                playsound(alarm_sound_path)
                closed_eyes_start_time = 0

    else:
        status = "focus"
        percentage = int(predictions * 100)
        closed_eyes_start_time = 0
        closed_eyes_duration = 0

    cv2.rectangle(frame, (0, 0), (200, 50), (255, 255, 255), -1)

    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, status, (10, 30), font, 1, (0, 0, 255), 2, cv2.LINE_4)
    cv2.putText(frame, f"{percentage}%", (10, 100), font, 1, (0, 0, 255), 2, cv2.LINE_4)
    cv2.putText(frame, f"{int(closed_eyes_duration)} seconds", (10, 200), font, 1, (0, 0, 255), 2, cv2.LINE_4)

    cv2.imshow("TukKantuk", frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
